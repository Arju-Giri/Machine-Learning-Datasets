```python
data.drop(columns = ["User_ID"], inplace = True)
```

```python
data["Gender"] = data["Gender"].map({
  "Male" : 0,
  "Female" : 1,
  "Non-binary" : 2
  })
```

```python
data["Platform"] = data["Platform"].astype("category").cat.codes
```

```python
data["Emotional_State_Post_Usage"] = data["Emotional_State_Post_Usage"].astype("category").cat.codes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

## Step 1

## Choose target variable (Y)

## Target = Addiction_Level

```python
X = data.drop(columns= ["Addiction_Level"])
y = data["Addiction_Level"]
```

## Step 2 ; split dataset (Most important step)


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

## Step 3 : Create ML Model (Decision Tree)

```python

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

## Step 4 : Make Predictions 

```python
y_pred = model.predict(X_test)
```

## Step 5 : Check accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy :", accuracy)
```
